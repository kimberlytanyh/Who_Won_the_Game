```python
import random

def bowling_game():
    #Part 1
    player_num = int(input("Enter the number of players: "))
    players = []
    
    for i in range(1, player_num+1):
        player_name = input("Enter player name: ")
        players.append(player_name)
    print("\n")
    
    #Part 2
    
    def bowling_round():
        roll1 = random.randint(0,10)
        if roll1 == 10:
            roll2 = str("/")
        elif roll1 < 10 and roll1 >= 0:
            roll2 = random.randint(0,10-roll1)
        return roll1, roll2

    #Part 3
    score_card = open('scorecard.txt','w')
    for round in range(1,11):
        print("Round",str(round)+":")
        for player in range(0, player_num):
            roll1, roll2 = bowling_round()
            print(players[player]+ ": ", roll1, roll2)
            score_card.write(players[player]+"\n"+str(roll1)+"\n"+str(roll2)+"\n")
        print("\n")
    score_card.close()
    
#Part 4
    def Totalscore(player):
        score_card = open('scorecard.txt','r')
        totalscore = 0
        playerName = score_card.readline().rstrip()
       
        while playerName!="":
            playername = playerName
            roll1 = score_card.readline().rstrip()
            roll2 = score_card.readline().rstrip()
            
            if player == playerName:
                totalscore += int(roll1)
                if roll2 != "/":
                    totalscore += int(roll2)
            playerName = score_card.readline().rstrip()
        score_card.close()
        return totalscore

    print("Final Score per Player: ")
    for i in range(0, player_num):
        player= players[i]
        print(str(player)+ "'s Score: " + str(Totalscore(player)))

bowling_game()
```
